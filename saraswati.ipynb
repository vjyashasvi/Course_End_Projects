```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Import data into Python environment

```python
df=pd.read_csv("Comcast_telecom_complaints_data.csv")
df.head()
```

```python
df.info()
```

```python
df
```

```python
df["Date_Index"]=df["Date_month_year"]+" "+df["Time"]
df["Date_Index"]=pd.to_datetime(df["Date_Index"])
df=df.set_index(df['Date_Index'])
```

```python
df
```

## 2. Provide the trend chart for the number of complaints at monthly granularity levels.

```python
f=df.groupby(pd.Grouper(freq='M')).size()
f
```

```python
ax=plt.axes()
ax.set_facecolor('black')
f.plot(figsize=(10,10),color='lime',linewidth=5)
plt.title(" Trend chart for the number of complaints at monthly granularity levels",size=20)
plt.xlabel('Months')
plt.ylabel('Frequency')
```

## 2. Provide the trend chart for the number of complaints at Daily granularity levels.

```python
ax=plt.axes()
ax.set_facecolor('black')
df.groupby(pd.Grouper(freq="D")).size().plot(figsize=(11,7),color='aqua',linewidth=2.5)

plt.title("Trend chart for the number of complaints at daily granularity levels",size=20,loc='left',color='red')
plt.xlabel('Days of Months')
plt.ylabel('Frequency')
```

4.Create a new categorical variable with value as Open and Closed. Open & Pending is to be categorized as Open and Closed & Solved is to be categorized as Closed.

```python
df['newStatus']=['open' if Status=='Open' or Status=='Pending' else 'closed' for Status in df['Status']]
```

```python
df
```

## 5. Provide state wise status of complaints in a stacked bar chart. Use the categorized variable from Q3.

```python
df_status = df.groupby('State').newStatus.value_counts().unstack()
df_status.fillna(0,inplace=True)
df_status
```

```python

df_status.plot(kind='bar',stacked=True,figsize=(17,10),color=['lawngreen','red']).set_facecolor('black')
plt.title('State wise Opened/Closed Complaints',size=40,color='teal')
```

## 7. Which state has the highest percentage of unresolved complaints

## 8. Provide the percentage of complaints resolved till date, which were received through the Internet and customer care calls.
