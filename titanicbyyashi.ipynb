```python
df['Title'] = df.Name.str.extract(' ([A-Za-z]+)\.', expand=False)
```

```python
df['Title'] = df['Title'].replace(['Lady', 'Countess','Capt', 'Col','Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona', 'Master'], 'Elite')
df['Title'] = df['Title'].replace(['Mlle','Miss'], 'Miss')
df['Title'] = df['Title'].replace(['Mme','Ms'], 'Mrs')
```

```python
title= {"Mr": 1, "Miss": 2, "Mrs": 3, "Royal": 5}
df['Title'] = df['Title'].map(title)
df.isnull().sum()
```

```python
df['Title'] = df['Title'].fillna(0)
df
```

```python
df=df.drop(['Name','Ticket','PassengerId'], axis=1)
df
```

```python
sex= {"male": 1, "female": 2}
df['Sex'] = df['Sex'].map(sex)
df
```

```python
df['Embarked'].unique()
```

```python
em= {"S": 1, "C": 2, 'Q': 3}
df['Embarked'] = df['Embarked'].map(em)
df
```

```python
df['Age']=df['Age'].fillna((df['Age'].mean()))
df['Embarked']=df['Embarked'].fillna(1)
df.isnull().sum()
```

```python
df.describe()
```

```python
df
```

```python
df.loc[ df['Age'] <= 16, 'Age'] = 0
df.loc[(df['Age'] > 16) & (df['Age'] <= 32), 'Age'] = 1
df.loc[(df['Age'] > 32) & (df['Age'] <= 48), 'Age'] = 2
df.loc[(df['Age'] > 48) & (df['Age'] <= 64), 'Age'] = 3
df.loc[ df['Age'] > 64, 'Age']
```

```python
df.groupby(['Survived','Age'])['Age'].size()
```

```python
std_scale_fare = StandardScaler()
std_scale_fare.fit(df[['Fare']])
```

```python
std_scale_fare.mean_, std_scale_fare.scale_
```

```python
df['Fare'] = std_scale_fare.transform(df[['Fare']])
df.columns
```

```python
df['SibSp'].unique()
```

```python
df.loc[(df['SibSp'] + df['Parch'] )==0, 'Family'] = 0
df.loc[(df['SibSp'] + df['Parch'] )>=1, 'Family'] = 1
```

```python
df
```

```python
df.columns
```

```python
features=df[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare',
       'Embarked', 'Title', 'Family']]
target=df[['Survived']]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score, classification_report
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, random_state=10)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
my_logreg_model = LogisticRegression().fit(X_train, y_train)
```

```python
my_logreg_preds_train = my_logreg_model.predict(X_train)
my_logreg_preds_test = my_logreg_model.predict(X_test)

print('Accuracy on Train set : ',accuracy_score(y_train, my_logreg_preds_train))
print('Accuracy on Test set` : ',accuracy_score(y_test, my_logreg_preds_test))
```

```python
list(zip(my_logreg_model.predict_proba(X_train), my_logreg_model.predict(X_train)))
```

```python
print('Confusion Matrix on Train set :\n', confusion_matrix(y_train, my_logreg_preds_train))
```

```python
print('Confusion Matrix on Test set :\n', confusion_matrix(y_test, my_logreg_preds_test))
```

```python
print('Precision on Train set : ',precision_score(y_train, my_logreg_preds_train))
print('Precision on Test set : ',precision_score(y_test, my_logreg_preds_test))

print('\n')

print('Recall on Train set : ',recall_score(y_train, my_logreg_preds_train))
print('Recall on Test set : ',recall_score(y_test, my_logreg_preds_test))

print('\n')

print('F1-Score on Train set : ',f1_score(y_train, my_logreg_preds_train))
print('F1-Score on Test set : ',f1_score(y_test, my_logreg_preds_test))
```

```python
print('Classification Report on Train set : ')
print(classification_report(y_train, my_logreg_preds_train))
```

```python
print('Classification Report on Test set : ')
print(classification_report(y_test, my_logreg_preds_test))
```
